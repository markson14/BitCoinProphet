```python
import pandas as pd
from fbprophet import Prophet
import numpy as np
```

```python
df = pd.read_csv('btcusd.csv')
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=700)
future.tail()
```

```python
forecast = m.predict(future)
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
